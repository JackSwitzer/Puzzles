```python
import math
import pandas as pd
import numpy as np
```

```python
def isPalindrome(num):
    return num == rev(num)
    
def rev(num):
    return int(num != 0) and ((num % 10) * \
            (10**int(math.log(num, 10))) + \
                        rev(num // 10))
```

```python
def isPerfectSquare(x):
    s = int(math.sqrt(x))
    return s*s == x
 
# Returns true if n is a Fibonacci Number, else false
 
 
def isFibonacci(n):
 
    # n is Fibonacci if one of 5*n*n + 4 or 5*n*n - 4 or both
    # is a perfect square
    return isPerfectSquare(5*n*n + 4) or isPerfectSquare(5*n*n - 4)
```

```python
def generate_palindromes_up_to_limit(limit):
    def create_palindrome(num_str, is_odd):
        return int(num_str + num_str[:-is_odd][::-1])

    limit_sqrt = int(limit ** 0.5)
    palindromes = []

    # Generate palindromes with even number of digits
    for i in range(1, limit_sqrt):
        num_str = str(i)
        palindrome = create_palindrome(num_str, False)
        if palindrome >= limit:
            break
        palindromes.append(palindrome)

    # Generate palindromes with odd number of digits
    for i in range(1, limit_sqrt):
        num_str = str(i)
        palindrome = create_palindrome(num_str, True)
        if palindrome >= limit:
            break
        palindromes.append(palindrome)

    return palindromes

limit = 10 ** 10
palindromes = generate_palindromes_up_to_limit(limit)
print(palindromes)

#modify this so we get all palindromes up to 1 trillion
#can use this to get the others 
#then also will be able to know possible plaindrome multiples of 23 for example
```

```python
def palindrome_number_generator():
    yield 0    
    lower = 1
    while True:
        higher = lower*10
        for i in xrange(lower, higher):    
            s = str(i)
            yield int(s+s[-2::-1])
        for i in xrange(lower, higher):    
            s = str(i)
            yield int(s+s[::-1])
        lower = higher


def palindromes(lower, upper):
    all_palindrome_numbers = palindrome_number_generator()
    for p in all_palindrome_numbers:
        if p >= lower:
            break
    palindrome_list = [p]
    for p in all_palindrome_numbers:
        # Because we use the same generator object,
        # p continues where the previous loop halted.
        if p >= upper:
            break
        palindrome_list.append(p)
    return palindrome_list


print palindromes(1, 10**10)

```

```python
iCount = [0] * 10
for i in range(0, 1000):
    temp = str(i*37)
    iCount[int(temp[-1])] += 1

print(iCount)
    
```

```python
iCount = [0] * 10
for i in range(0, 1000):
    temp = str(i*88)
    iCount[int(temp[-1])] += 1

print(iCount)
    
```

```python
tempList = [[]]
```

```python
for num in range(0, 10**9):
    insertRow = [False, False, False, False]
    
    insertRow[0] = isPalindrome(num)
    insertRow[1] = isFibonacci(num)
    insertRow[2] = num % 37 == 0
    insertRow[3] = num % 88 == 0

    tempList.append(insertRow)

print(tempList)
df = pd.DataFrame(tempList, columns=['Palindrome', 'Fibonacci', 'Multiple of 37', 'Multiple of 88'])
df = df.shift(-1)


df.to_csv('JSMay2024Data.csv')
```
